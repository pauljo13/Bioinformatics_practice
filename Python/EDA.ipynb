```python
gs.loc[gs['ID'] == epc[1]]['Gene Symbol'].isna
gs.loc[gs['ID'] == epc[1], 'Gene Symbol']
print(gs[gs['ID'] == epc[1]]['Gene Symbol'].values) 
```

```python
New_col = []
for i in epc:
    # 특정 ID와 매칭되는 'Gene Symbol' 찾기
    matched_gene = gs.loc[gs['ID'] == i, 'Gene Symbol']

    if matched_gene.isna().any():  
        New_col.append(i)  # NaN이면 원래 ID를 사용
    elif not matched_gene.empty:  # 매칭되는 값이 있을 경우
        New_col.append(matched_gene.iloc[0])  # 단일 값 추가
    else:
        New_col.append(i)  # ID가 없을 경우 원래 값 유지

print(New_col)
```
